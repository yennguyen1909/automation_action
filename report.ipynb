Trích các cặp hành động để làm Association Rule Mining (A → B)

```python
from collections import Counter
```

```python
pair_counter = Counter()

for seq in sequences['action']:
    actions = [x.strip() for x in seq.split('->')]
    for i in range(len(actions) - 1):
        pair = (actions[i], actions[i + 1])
        pair_counter[pair] += 1

pair_df = pd.DataFrame(pair_counter.items(), columns=['pair', 'count'])
pair_df[['from_action', 'to_action']] = pd.DataFrame(pair_df['pair'].tolist(), index=pair_df.index)
pair_df = pair_df.drop(columns=['pair']).sort_values(by='count', ascending=False)

print(pair_df.head(10))
```

```python
for seq, count in top_sequences:
    steps = seq.split(" -> ")
    for i in range(len(steps) - 1):
        source, target = steps[i], steps[i + 1]
        links[(source, target)] += count

        for step in (source, target):
            if step not in label_to_index:
                label_to_index[step] = len(node_labels)
                node_labels.append(step)
```

```python
node_labels = []
label_to_index = {}
links = defaultdict(int)
```

```python

for seq, count in top_sequences:
    steps = seq.split(" -> ")
    for i in range(len(steps) - 1):
        source, target = steps[i], steps[i + 1]
        links[(source, target)] += count

        for step in (source, target):
            if step not in label_to_index:
                label_to_index[step] = len(node_labels)
                node_labels.append(step)

# Chuẩn bị dữ liệu cho Sankey diagram
source_indices = []
target_indices = []
values = []

for (source, target), value in links.items():
    source_indices.append(label_to_index[source])
    target_indices.append(label_to_index[target])
    values.append(value)

# Vẽ biểu đồ Sankey
sankey_fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=node_labels
    ),
    link=dict(
        source=source_indices,
        target=target_indices,
        value=values
    )
)])
sankey_fig.update_layout(title_text="Sankey Diagram với luồng chuỗi hành động phổ biến", font_size=10)
```
